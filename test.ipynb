```python
from pyswip import Prolog
pro = Prolog()
```

```python
ndt = pd.read_csv("ndt_all_reduce.csv", index_col = 0)
```

```python
def get_sentence(ident):
    d = ndt[ndt.sent_id == ident]
    for i in d["token_order head".split()].iterrows():
        prolog.assertz(f"dep({ident},{i[1][0]}, {i[1][1]})")
    return True 
```

```python
def f(x):
    try:
        x = x.replace("'", '').replace('"', '')
    except:
        x = x
    return x          
```

```python
ndt.sample(1)
```

```python
def dump_ndt_prolog():
    for i in ndt.iterrows():
        r = i[1]
        pro.assertz(f"""ndt({r[0]},{r[1]},'{f(r[2])}','{f(r[3])}','{f(r[4])}','{f(r[5])}',{r[6]},'{r[7]}',{r[8]},{r[9]},'{f(r[10])}')""") 
```

```python
def write_ndt_prolog(filename):
    with open(filename, "w", encoding = "utf-8") as fp:
        for i in ndt.iterrows():
            r = i[1]
            clause = f"""ndt({r[0]},{r[1]},'{f(r[2])}','{f(r[3])}','{f(r[4])}','{f(r[5])}',{r[6]},'{r[7]}',{r[8]},{r[9]},'{f(r[10])}')."""
            fp.write(clause)
            fp.write("\n")
```

```python
write_ndt_prolog('ndt_all.pl')
```

```python
dump_ndt_prolog()
```
